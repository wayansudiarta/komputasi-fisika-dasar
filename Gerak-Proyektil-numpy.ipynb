Menggunakan metode numerik Euler.
$$ x_{n+1} = x_n + v_{x,n}\Delta t$$
$$ y_{n+1} = y_n + v_{y,n}\Delta t$$
$$ v_{x,n+1} = x_n + a_{x,n}\Delta t$$
$$ v_{y,n+1} = x_n + a_{y,n}\Delta t$$
dengan $a_{x,n} = 0$ dan $a_{y,n} = -g$


```python
import numpy as np
```

```python
N = 100
# Array
x = np.zeros(N+1)
vx = np.zeros(N+1)
y = np.zeros(N+1)
vy = np.zeros(N+1)
```

```python
# Parameter
theta = 30.0
v0 = 10.0
vx[0] = v0*np.cos(theta*np.pi/180)
vy[0] = v0*np.sin(theta*np.pi/180)
y[0] = 5 
x[0] = 0
ax = 0
ay = -9.8
dt = 0.02
```

```python
t = np.arange(0,N+1)*dt
```

```python
# Iterasi
for n in range(N):
    x[n+1] = x[n] + vx[n]*dt 
    vx[n+1] = vx[n] + ax*dt
    y[n+1] = y[n] + vy[n]*dt
    vy[n+1] = vy[n] + ay*dt
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y,'ob')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Gerak Proyektil')

plt.hlines(0, 0, 20)   # Tanah
plt.hlines(5, -1, 0)   # gedung
plt.vlines(0, 0, 5)    # 
plt.xlim(-1, 20)    # reset plot limits
plt.show()
```
